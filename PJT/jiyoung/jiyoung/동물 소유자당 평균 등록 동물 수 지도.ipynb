```python
import numpy as np
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import json
```

```python
fm = mp.font_manager.FontManager()
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df_animal = pd.read_csv('data_crowing/반려동물등록현황.csv', encoding='cp949', engine='python' )
df_animal.head()
```

```python
df_animal.info()
```

```python
df_animal['시군명'].unique()
```

```python
df = df_animal[df_animal['시군명']!='가평군'].copy()
df['시군명'].unique()
```

```python
dong_df = df[['시군명','동물소유자당등록동물수']]
dong_df
```

### 지역 합치기

```python
new_df = dong_df.groupby(dong_df['시군명']).sum()
new_df
```

### 동물소유자당등록동물수 평균

```python
new_df = round(dong_df.groupby(dong_df['시군명']).mean(),2)
new_df
```

```python
df_sigun_avg = new_df.groupby(['시군명'])['동물소유자당등록동물수'].mean()
print(df_sigun_avg)
print(df_sigun_avg.index)
print(df_sigun_avg.values)
new_df_sigun_avg = pd.DataFrame({'시군명':df_sigun_avg.index, '등록동물수':df_sigun_avg.values})
print(new_df_sigun_avg.info())
new_df_sigun_avg
```

```python
new_df_sigun_avg.head()
```

### 그래프 시각화

```python
new_df_sigun_avg.plot()
plt.figure(figsize=(15,4))
plt.xticks(rotation=45)
sns.lineplot(data=new_df_sigun_avg, x ='시군명', y ='등록동물수')
```

```python
new_df_sigun_avg[['시군명', '등록동물수']].plot.scatter(x='시군명', y='등록동물수', figsize=(15,4), rot=45, grid=True) 
```

### 지도 시각화

```python
geo_path = 'C:/Users/82108/study/202107-반려동물-프로젝트/프로젝트_자료/data_crowing/GyeonggiSiGun.geojson'
geo_str = json.load(open(geo_path, encoding='utf-8'))
geo_str
```

```python
#지도 기본 출력 설정: 경기도 중심부(하남시) 위/경도

map = folium.Map(location=[37.540393, 127.164423], 
                 zoomm_start=5, tiles='Stamen Terrain')


#columns = ('geo_data랑 연결할 컬럼','표시하고싶은 컬럼)
#key_on='feature.properties.sggnm'  <- sggnm 자리에 data랑 연결할 거, 여기서는 시군(geo_data의 json코드에서 찾기)
folium.Choropleth(
    geo_data = geo_str,
    data = new_df_sigun_avg,
    columns = ('시군명','등록동물수'),
    fill_color='BuPu',
    key_on='feature.properties.sggnm',
    legend_name='지역내 등록동물수 평균'
).add_to(map)

map
```

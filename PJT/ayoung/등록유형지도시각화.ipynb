# 데이터 전처리

```python
df_kosis['시군명'].unique()
```

```python
df=df_kosis.iloc[:,[0,6,7,8]]
df.head()
```

```python
new_df=df.groupby(df['시군명']).sum()
```

```python
new_df
```

```python
new_dfper=new_df.assign(per_in=lambda x:(x['(RFID종류)내장형']/(x['(RFID종류)내장형']+x['(RFID종류)외장형']+x['(RFID종류)인식표']))*100,
                      per_ex=lambda x:(x['(RFID종류)외장형']/(x['(RFID종류)내장형']+x['(RFID종류)외장형']+x['(RFID종류)인식표']))*100,
                      per_tag=lambda x:(x['(RFID종류)인식표']/(x['(RFID종류)내장형']+x['(RFID종류)외장형']+x['(RFID종류)인식표']))*100)
new_dfper
```

```python
per_df=new_dfper.loc[:,'per_in':'per_tag']
per_df.head()
```

```python
per_df.to_csv("RFID비율.csv")
```
